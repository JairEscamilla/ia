```python
# Importando librerias
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
```

```python
# Cargamos la libreria
boston = load_boston()

print(boston.DESCR)
```

```python
X = np.array(boston.data[:, 5])
Y = np.array(boston.target)

plt.scatter(X, Y, alpha = 0.3)
plt.show()
```
